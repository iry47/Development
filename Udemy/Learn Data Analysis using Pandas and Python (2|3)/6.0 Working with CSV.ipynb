```python
data = pd.read_csv('/Users/ryanheadley/Documents/GitHub/Development/Udemy/Learn Data Analysis using Pandas and Python (2|3)/ressources/_occupancyrates.csv',index_col="Time")
data.head()
```

```python
data.shift(1).head()
```

#### Every row was shifted to the following time period (the next month)

```python
data['Previous Total Occupancy'] = data['Total Occupancy rate (percent)'].shift(1)
data.head()
```

```python
data['Diff in Occupancy from yesterday'] = data['Total Occupancy rate (percent)'] - data['Previous Total Occupancy']
data.head()
```
